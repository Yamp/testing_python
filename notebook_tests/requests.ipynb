# Различные примеры есть здесь https://httpbin.org

# GET

```python
import requests, pprint
r = requests.get('https://httpbin.org/get', params = {'key1':'value1', 'key2':'Вова'}) 
print(r)
pprint.pprint(r.text)
# pprint.pprint(eval(r.text))
```

# PUT

```python
import requests
r = requests.put('https://httpbin.org/put', params = {'key1':'value1', 'key2':'Вова'}) 
print(r.text)
```

# Получить xml из ОПТа

```python
import requests, pprint
r = requests.get('http://srv-bl/GetXml', params = {'data' : 'TovInfo', 'anyCode' : 141810, 'nakl_id' : 0}) 
pprint.pprint(r.text)
print("Кодировка", r.encoding)
```

# json

```python
import requests, pprint  
r = requests.get('https://httpbin.org/json')  
pprint.pprint(r.json())
```

# Более сложные POST запросы (json)

```python
import requests, pprint
payload_dict = {'slideshow': {'author': 'Yours Truly',
               'date': 'date of publication',
               'slides': [{'title': 'Wake up to WonderWidgets!', 'type': 'all'},
                          {'items': ['Why <em>WonderWidgets</em> are great',
                                     'Who <em>buys</em> WonderWidgets'],
                           'title': 'Overview',
                           'type': 'all'}],
               'title': 'Sample Slide Show'}}
r2 = requests.post('https://httpbin.org/post', json=payload_dict)
print(r2.text)  
```

# Передать файлы на сервер POST запросом

# cookies
